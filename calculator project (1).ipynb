```python
# This function is for addition of two numbers 
def addition(x, y):
    return x + y

# This function is for subtraction of two numbers
def subtraction(x, y):
    return x - y

# This function is for multiplication of two numbers
def multiplication(x, y):
    return x*y

# This function is for division of two numbers
def division(x, y):
    return x/y

print ("select operation")
print("1.addition")
print("2.subtraction")
print("3.multiplication")
print("4.division")

while True:
    # take input from the user
    choice = input("Enter choice (1 or 2 or 3 or 4) : ")
    
    if choice in ('1', '2', '3', '4'):
        try:
            number_1 = float(input("Enter 1st number : "))
            number_2 = float(input("Enter 2nd number : "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
            
        if choice == '1':
            print(number_1, "+", number_2, "=", addition(number_1, number_2))
            
        elif choice == '2':
            print(number_1, "-", number_2, "=", subtraction(number_1, number_2))
            
        elif choice == '3':
            print(number_1, "*", number_2, "=", multiplication(number_1, number_2))
            
        elif choice == '4':
            print(number_1, "/", number_2, "=", division(number_1, number_2))
            
            
        # check if user wants another calculation
        # break the while loop if answer is no
        next_calculation = input("Want to do next calculation? (yes/no): ")
        if next_calculation == "no":
            break
    
    else:
        print("Invalid Input")
    
```

# Basic calculator with square and cube 

```python
# This function is for addition of two numbers 
def addition(x, y):
    return x + y

# This function is for subtraction of two numbers
def subtraction(x, y):
    return x - y

# This function is for multiplication of two numbers
def multiplication(x, y):
    return x*y

# This function is for division of two numbers
def division(x, y):
    return x/y

# This function is for square of number
def Square(num):
    return num**2

# This function is for cube of number
def Cube(num):
    return num**3

print ("select operation")
print("1.addition")
print("2.subtraction")
print("3.multiplication")
print("4.division")
print("5.square")
print("6.cube")

while True:
    choice = input("Enter choice (1 or 2 or 3 or 4 or 5 or 6): ")
    
    if choice in ('1', '2', '3', '4'):
        try:
            number_1 = float(input("Enter 1st number: "))
            number_2 = float(input("Enter 2nd number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        
        if choice == '1':
            print(number_1, "+", number_2, "=", addition(number_1, number_2))
        elif choice == '2':
            print(number_1, "-", number_2, "=", subtraction(number_1, number_2))
        elif choice == '3':
            print(number_1, "*", number_2, "=", multiplication(number_1, number_2))
        elif choice == '4':
            if number_2 == 0:
                print("Division by zero is not allowed!")
            else:
                print(number_1, "/", number_2, "=", division(number_1, number_2))
    
    elif choice in ('5', '6'):
        try:
            num = float(input("Enter a number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        
        if choice == '5':
            print("Square of the number", num, "is", Square(num))
        elif choice == '6':
            print("Cube of the number", num, "is", Cube(num))
    
    else:
        print("Invalid Input")
    
    next_calculation = input("Want to do next calculation? (yes/no): ")
    if next_calculation.lower() != "yes":
        break
```
